```python
final = []
with open('./Datasets/NSA Snort/alert.full') as f:
    tmp = ""
    lines = f.readlines()
    for line in lines:
        if line == "\n":
            tmp = tmp.replace('\n', '')[:-1]
            final.append(tmp)
            tmp = ""
            continue
        tmp = tmp + line + " "
with open('./Datasets/NSA Snort/alert_new.full', 'w') as fw:
    for line in final:
        fw.write(line+'\n')
```

```python
final = []
with open("./Datasets/NSA Snort/alert_new.full") as file:
    content = file.readlines()
    for line in content:
        try:
            all_vals = re.findall(r'^.*?\[Classification:\s+(.*?)\]\s+\[Priority:\s+(.*?)\]\s+\d+\/\d+-\d+:\d+:\d+.\d+\s+(\d+.\d+.\d+.\d+:\d+).*?(\d+.\d+.\d+.\d+:\d+)\s+(\w+)\s+TTL:(\d+)\s+TOS:(\w+)\s+ID:(\w+)\s+IpLen:(\d+)\s+DgmLen:(\d+)\s+DF\s+(.+?)\s+Seq:\s+(\w+)\s+Ack:\s+(\w+).*?TcpLen:\s+(\d+)', line)
        #all_vals = [str(x) for x in all_vals]
        #print(all_vals[0], type(all_vals[0]))
            final.append(','.join(list(all_vals[0])))
        except:
            pass
```
