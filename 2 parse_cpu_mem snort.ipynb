### Code to parse snort_cpu_mem.log

```python
import re

d = {}
with open('Datasets/Snort CPU Memory Log/snort_cpu_mem.log') as f:
    lines = f.readlines()
    lines = lines[1:]
    tmp = []
    for line in lines:
        #print(line)
        line = line.strip()
        if not line.startswith(r"%CPU"):
            tmp.append(line)
        
        else:
            tokens = line.split()
            m = tokens[7]
            if((int(m.split(":")[0]) == 11 and int(m.split(":")[1]) >= 40) or \
                (int(m.split(":")[0]) == 12 and int(m.split(":")[1]) <=10)):
                d[m] = tmp
            tmp=[]


            
```

```python
def getVals(lst):
    #lst[-1].replace("\n")
    for line in lst:
        line = line.strip()
        tokens = line.split()
        #print(tokens)
        if(tokens[len(tokens)-2]=="snort" and tokens[len(tokens)-1] == "-A") or  tokens[len(tokens)-1]=="snort":
            #print("hre")
            return [float(tokens[0]), float(tokens[1])]
```

```python
for key, value in d.items():
    #print(key,value)
    d[key] = getVals(value)
    
    
```

```python
s_c, s_m = 0,0

for k,v in d.items():
    s_c+=v[0]
    s_m+=v[1]
    
```
