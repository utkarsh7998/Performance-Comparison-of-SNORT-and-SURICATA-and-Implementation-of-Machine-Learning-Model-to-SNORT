### Code to parse ps.log

```python
import re

d = {}
with open('Datasets/Suricata CPU Memory Log/ps.log') as f:
    lines = f.readlines()
    lines = lines[1:]
    tmp = []
    for line in lines:
        line = line.strip()
        if not line.startswith(r"%CPU"):
            tmp.append(line)
        #print(line)
        else:
            tokens = line.split()
            m = tokens[7]
            if(int(m.split(":")[0]) == 11 and int(m.split(":")[1]) >= 21 and int(m.split(":")[1]) <=45):
                d[m] = tmp
            tmp=[]


            
```

```python
def getVals(lst):
    #lst[-1].replace("\n")
    for line in lst:
        line = line.strip()
        tokens = line.split()
        #print(tokens)
        if(tokens[len(tokens)-2]=="/usr/bin/suricata" and tokens[len(tokens)-1] == "-c") :
            #print("hre")
            return [float(tokens[0]), float(tokens[1])]
```

```python
for key, value in d.items():
    d[key] =  getVals(value)
    
    
```

```python
s_c, s_m = 0,0

for k,v in d.items():
    s_c+=v[0]
    s_m+=v[1]
    
```
